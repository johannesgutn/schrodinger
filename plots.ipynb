```python
df50 = df[df.grid_points == 50]
df50.sort_values(['','z','E']).head(40)
```

```python
df60 = df[df.grid_points == 60]
df60.sort_values(['','z','E']).head(60)
```

```python
40 = np.array(df40[''])
t_failed40 = np.array(df40['t_failed'])

50 = np.array(df50[''])
t_failed50 = np.array(df50['t_failed'])

60 = np.array(df60[''])
t_failed60 = np.array(df60['t_failed'])


plt.semilogx(40,t_failed40)
plt.semilogx(50,t_failed50)
plt.semilogx(60,t_failed60)
plt.show()
```
