```python
import matplotlib.pyplot as plt
import numpy as np

all = np.load('data_files/sch_p=0.1_theta=0.1_gridpoints=60_gridsize=3_L=0.5_E=5_z=0.4.npy')

t = all[0]
full1 = all[1]
full2 = all[2]
Nc1 = all[3]
Nc2 = all[4]
Nc1_true = all[5]

```

```python
plt.plot(t,np.real(full1), label='K1 Numeric-full-re',color='C0')
plt.plot(t,np.imag(full1), label='K1 Numeric-full-im',linestyle='--',color='C0')

plt.plot(t,np.real(full2), label='K2 Numeric-full-re',color='C1')
plt.plot(t,np.imag(full2), label='K2 Numeric-full-im',linestyle='--',color='C1')


plt.legend()
plt.show()
```

```python
plt.plot(t,np.real(Nc1),label='Nc K1 real',linestyle='--',color='C0')
plt.plot(t,np.real(Nc1_true),label='Nc K1 real true value',linestyle=':',color='C0')
plt.plot(t,np.real(full1),label='Full K1 real',color='C0')

plt.plot(t,np.imag(Nc1),label='Nc K1 imag',linestyle='--',color='C1')
plt.plot(t,np.imag(Nc1_true),label='Nc K1 imag true value',linestyle=':',color='C1')
plt.plot(t,np.imag(full1),label='Full K1 imag',color='C1')

plt.legend()
plt.show()
```

```python
plt.plot(t,np.real(Nc2),label='Nc K2 real',linestyle='--',color='C0')
plt.plot(t,np.real(full2),label='Full K2 real',color='C0')

plt.plot(t,np.imag(Nc2),label='Nc K2 imag',linestyle='--',color='C1')
plt.plot(t,np.imag(full2),label='Full K2 imag',color='C1')

plt.legend()
plt.show()
```
